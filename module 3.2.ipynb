```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
URL = 'https://index.minfin.com.ua/ua/russian-invading/casualties/'

def get_url_list():
    html_doc = requests.get(URL)
    soup = BeautifulSoup(html_doc.text, 'html.parser')
    pref = 'month.php?month='
    url_list = []

    content = soup.select('div[class=ajaxmonth] h4[class=normal] a')
    for link in content:
        url = pref + re.search(r"\d{4}-\d{2}", link['href']).group()
        url_list.append(url)

    return url_list
```

```python
get_url_list()
```

```python
source = 'https://index.minfin.com.ua/ua/russian-invading/casualties/month.php?month=2024-07'
```

```python
all = {}

for link in get_url_list():

    monthly_key = link[-7:]
    html_doc = requests.get(URL + link)
    soup = BeautifulSoup(html_doc.text, 'html.parser')
    content = soup.select('ul[class=see-also] li[class=gold]')

    dict_2 = {}

    for li in content:
        data_key = li.find('span', attrs={'class': 'black'}).text
        casualties = li.find('div').find('div').find('ul')

        dict_3 = {}

        for t in casualties:
            name, quantities = t.text.split('—')
            name = name.strip()
            quantities = re.search(r"\d+", quantities).group()
            dict_3[name] = quantities

        dict_2[data_key] = dict_3

    all[monthly_key] = dict_2


all
```

```python
import json

with open('vtraty.json', 'w', encoding='utf=8') as f:
    json.dump(all, f, ensure_ascii=False, indent=4)
```
